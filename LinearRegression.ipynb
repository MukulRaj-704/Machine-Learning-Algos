Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

UNDERSTANDING LINEAR REGRESSION

```python
x=10*np.random.rand(100,1)
y= 4+3*x+np.random.randn(100,1)
```

```python
sns.scatterplot(x=x.flatten(), y=y.flatten())
plt.xlabel("input value")
plt.ylabel("output value")
plt.title("scatter plot between x and y")
```

```python
x_b = np.c_[np.ones((100,1)),x]
```

```python
theta = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
theta
```

```python
x_new = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
new_b = np.c_[np.ones((100,1)), x]
y_pred = new_b.dot(theta)
y_pred
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
x_new = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
y_predict = lr.predict(x)
```

```python
print('intercept',lr.intercept_)
print('slop',lr.coef_)
```

Comparing both Model Graph

```python
plt.scatter(x,y,color='blue',label='Data Points')
plt.plot(x,y_pred,color='red',label='Normal-Form Prediction')
plt.plot(x, y_predict, color='green', label="SK-Lrearn prediction",linestyle='dotted')
plt.xlabel('Input Feature')
plt.ylabel('Output Feature')
plt.title("Comaprison between both Model")
plt.legend()
```

Multiple Linear Regression model

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
np.random.seed(42)
x_multi = np.random.rand(100, 5)
theta_true = np.array([[5], [2], [3], [1], [4], [0.5]])
x_b_multi = np.c_[np.ones((100, 1)), x_multi]

y_multi = x_b_multi.dot(theta_true) + np.random.randn(100, 1)

theta_manual = np.linalg.inv(x_b_multi.T.dot(x_b_multi)).dot(x_b_multi.T).dot(y_multi)
y_pred_manual = x_b_multi.dot(theta_manual)



```

```python
theta_true
```

```python
lrm = LinearRegression()
lrm.fit(x_multi, y_multi)
y_pred_sklearn = lrm.predict(x_multi)

y_multi_flat = y_multi.ravel()
y_pred_manual_flat = y_pred_manual.ravel()
y_pred_sklearn_flat = y_pred_sklearn.ravel()
```

```python
print(lrm.intercept_)
```

```python
print("Coefficients (weights):", lrm.coef_)

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_multi_flat, y_pred_manual_flat, color='blue', alpha=0.6, label="Manual Model")
plt.scatter(y_multi_flat, y_pred_sklearn_flat, color='green', alpha=0.6, label="Sklearn Model")
plt.plot([y_multi.min(), y_multi.max()], [y_multi.min(), y_multi.max()], 'r--', label='Ideal Fit')
plt.plot([y_multi.min(), y_multi.max()],[y_multi.min(), y_multi.max()])
plt.xlabel("Actual Y")
plt.ylabel("Predicted Y")
plt.title("Actual vs Predicted (Manual vs Sklearn)")
plt.legend()
plt.show()
```
