```python
import numpy as np
import pandas as pd
```

Creating a Random Dataset to Implement the Linear Regression

```python
np.random.seed(42)
temp=np.random.uniform(20,40,100)
sales=np.random.uniform(10000,43000,100)
```

It is a dataset in which we have temp and the sales and we are finding the linear relationship between them

```python
print(temp.shape)
print(sales.shape)
```

Mean Value of Temp And Sales

```python
temp_mean=np.mean(temp)
sales_mean=np.mean(sales)
```

```python
print(f"{temp_mean}\n{sales_mean}")
```

Slope Of the Equation y=mx+b

```python
slope=np.sum((temp-temp_mean)*(sales-sales_mean))/np.sum((temp-temp_mean)**2)
print(slope)
```

Intercept of the equation

```python
intercept=sales_mean-slope*temp_mean
print(intercept)
```

Generating Test Data

```python
np.random.seed(42)
test_temp=np.random.uniform(20,40,20)
test_sales=np.random.uniform(10000,43000,20)
```

```python
pred_sales=slope*test_temp+intercept
print(pred_sales)
```

Calculating Mean Squared Error

```python
MSE = np.sum((test_sales - pred_sales)**2) / len(test_sales)
print(MSE)
```

Root Mean Squared Error

```python
RMSE = np.sqrt(np.sum((test_sales - pred_sales)**2) / len(test_sales))
print(RMSE)
```

Mean Absolute Error

```python
MAE = np.sum(np.abs(test_sales - pred_sales)) / len(test_sales)
print(MAE)
```

Implementing Linear Regression Using SkLearn

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,root_mean_squared_error
```

```python
LR=LinearRegression()
LR.fit(temp.reshape(-1,1),sales.reshape(-1,1))
```

```python
Sales_LR =LR.predict(test_temp.reshape(-1,1))
```

```python
print('intercept',LR.intercept_)
print('slope',LR.coef_)
```

```python
LR_MSE =mean_squared_error(test_sales,Sales_LR)
print(LR_MSE)
LR_RMSE=root_mean_squared_error(test_sales,Sales_LR)
print(LR_RMSE)
LR_MAE=mean_absolute_error(test_sales,Sales_LR)
print(LR_MAE)
R2=r2_score(test_sales,Sales_LR)
print(R2)
```
