```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
# axis level function
```

```python
# figure level function
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='time',style='sex',kind='scatter',size='size')
```

```python
gap=px.data.gapminder()
gap
```

```python
india=gap[gap['country']=='India']
india
```

```python
sns.lineplot(data=india,x='lifeExp',y='year')
```

```python
sns.relplot(data=india,x='lifeExp',y='year',kind='line')
```

```python
#facet plot -- work only with figure level function
sns.relplot(kind='scatter',data=tips, x='total_bill',y='tip',row='sex',col='smoker')
```
