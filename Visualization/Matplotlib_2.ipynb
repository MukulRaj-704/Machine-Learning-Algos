```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('/content/iris.csv')
iris
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.show()
```

```python
iris.sample(5)
```

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
plt.show()
plt.figure(figsize=(10,6))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='summer')
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
plt.show()
```

```python
df=pd.read_csv('/content/batter.csv')
df
```

```python

sample_df=df.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.figure(figsize=(15,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'])

for i in range(25):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(15,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.axhline(125,color='red')
plt.axvline(28,color='green')
for i in range(25):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
fig = plt.figure(figsize=(10,6))

# Subplot 1: Avg vs Strike Rate
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(df['avg'], df['strike_rate'])
ax1.set_title("Avg vs Strike Rate")

# Subplot 2: Batter vs Strike Rate (using numeric x)
ax2 = fig.add_subplot(2,2,2)
x = range(len(df['batter']))
ax2.scatter(x, df['strike_rate'])
ax2.set_xticks(x)
ax2.set_xticklabels(df['batter'], rotation=90)
ax2.set_title("Batter vs Strike Rate")

# Subplot 4 (empty or for future use)
ax4 = fig.add_subplot(2,2,4)

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(df['avg'], df['strike_rate'])
ax1.set_title("Avg vs Strike Rate")

ax2 = fig.add_subplot(2,2,2)
ax2.plot(df['avg'])
ax2.set_title("Average")

ax3 = fig.add_subplot(2,2,4)
ax3.hist(df['strike_rate'])
ax3.set_title("Strike Rate Distribution")

plt.tight_layout()
plt.show()
```
