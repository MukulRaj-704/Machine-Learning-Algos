```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
batsman=pd.read_csv('/content/sharma-kohli.csv')
```

```python
batsman
```

This line plot shows how runs change over different years. The ups and downs in the line indicate performance variation over time.

```python
plt.plot(batsman["index"],batsman['V Kohli'])
plt.grid()
```

This plot compares the yearly runs of two players over the same time period.

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title(' Kohli vs Sharma')
plt.show()
```

Line styles help differentiate multiple datasets when colors alone are not enough.

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=3)
plt.plot(batsman['index'],batsman['V Kohli'],color='red', linestyle='dashed', linewidth=.5)
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title(' Kohli vs Sharma')
plt.show()
```

Markers highlight individual data points along with trends.

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=3,marker='D',markersize=3)
plt.plot(batsman['index'],batsman['V Kohli'],color='red', linestyle='dashed', linewidth=.5,marker='+',markersize=15)
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title(' Kohli vs Sharma')
plt.show()
```

Legend clearly identifies which line belongs to which player.

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=3,marker='D',markersize=3,label='Rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='red', linestyle='dashed', linewidth=.5,marker='+',markersize=15,label='Virat')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title(' Kohli vs Sharma')
plt.legend()
plt.show()
```

```python
import numpy as np

years = ['2022', '2023', '2024']
rohit = [800, 900, 850]
virat = [820, 880, 920]

x = np.arange(len(years))
width = 0.35

plt.bar(x - width/2, rohit, width, label='Rohit')
plt.bar(x + width/2, virat, width, label='Virat')

plt.xticks(x, years)
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Grouped Bar Chart")
plt.legend()
plt.show()
```

```python
matches = ['Match1', 'Match2', 'Match3']
rohit = [45, 60, 30]
virat = [55, 40, 70]

plt.bar(matches, rohit, label='Rohit')
plt.bar(matches, virat, bottom=rohit, label='Virat')

plt.xlabel("Matches")
plt.ylabel("Runs")
plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```
