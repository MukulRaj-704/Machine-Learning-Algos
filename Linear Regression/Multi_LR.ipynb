```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

Creating a random dataset including the features tempreatre, humidity and price and analysing effect of these things on the sales of a ice cream

```python
np.random.seed(1)
temp=np.random.uniform(20,40,150).reshape(-1,1)
humidity = np.random.uniform(0,100,150).reshape(-1,1)
price = np.random.uniform(50,250,150).reshape(-1,1)
```

```python
print(temp)
```

creste the sales of item using a balanced equation

```python
noise = np.random.normal(0, 15, 150).reshape(-1,1)
sales_mod = (
    200
    +6*temp
    -0.8*price
    -0.6*humidity
    +noise
)
sales_mod = np.clip(sales_mod, 0, None)
```

```python
print(sales_mod)
```

```python
print(f"temp = {temp} \n humidity= {humidity} \n price={price}")
```

```python
X=np.hstack((temp,humidity,price))
```

```python
ones=np.ones((X.shape[0],1))
```

```python

X_b = np.hstack((ones,X))
```

```python
X_b_T=X_b.T
```

```python
B=np.linalg.inv((X_b_T@X_b))@(X_b_T@sales_mod)
```

```python
print(B)
```

```python
np.random.seed(42)
temp_test=np.random.uniform(20,40,30).reshape(-1,1)
humidity_test=np.random.uniform(0,100,30).reshape(-1,1)
price_test=np.random.uniform(50,250,30).reshape(-1,1)
```

```python
noise_test = np.random.normal(0, 15, 30).reshape(-1,1)

sales_test = (
    200
    + 6 * temp_test
    - 0.8 * price_test
    - 0.6 * humidity_test
    + noise_test
)
```

```python
sales_pred= (B[0]+B[1]*temp_test+B[2]*humidity_test+B[3]*price_test)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
RMSE=root_mean_squared_error(sales_test,sales_pred)
```
