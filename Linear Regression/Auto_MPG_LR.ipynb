Calculating Correlation

```python
correlation=Auto_MPG.corr(numeric_only=True)
print(correlation)
```

From the correlation matrix, cylinders, displacement, and weight are highly correlated, indicating that they contain similar information. Therefore, only one of these features should be selected for model training.

```python
df = Auto_MPG[['mpg','weight','acceleration','model year','origin']]
```

```python
df.head()
```

isnull() checks for missing values in a dataset and returns a Boolean output (True for null, False for non-null), helping identify missing data.

```python
df.isnull().sum()
```

describe() generates descriptive statistics of numerical features, including count, mean, standard deviation, minimum, maximum, and quartiles, helping summarize the data distribution.

```python
df.describe()
```

```python
x=df[['weight','acceleration','model year','origin']]
y=df[['mpg']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train.head()
```

Feature standardization is performed to ensure that all features contribute equally to the training process, preventing dominance by features with larger numerical scales.

Fitting the scaler on test data causes data leakage and leads to overly optimistic model performance. we dont use fit_transform() on test data.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
X_train
```

fit_transform() calculate the mean and sd and transform the data from dataframe to numpy array

transform() used in test data doen not calculate the mean and sd it uses the above calculated mean and sd

```python
from sklearn.linear_model import LinearRegression
LR_model = LinearRegression()
LR_model.fit(X_train, y_train)
```

```python
y_pred=LR_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print("R Score:", r2_score(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
coef = pd.DataFrame({
    'Feature': x_train.columns,
    'Coefficient': LR_model.coef_.flatten()
})
coef
```

R Score indicate that model explains 84% of the variance in MPG
The RMSE of 2.91 suggests that the models predictions deviate from actual MPG values by approximately 3 MPG on average.
